```python
def valid_password(password):
    has_upper = any(ch.isupper() for ch in password)
    has_digit = any(ch.isdigit() for ch in password)
    has_special = any(not ch.isalnum() for ch in password)
    long_enough = len(password) >= 8
    conditions_met = sum([has_upper, has_digit, has_special, long_enough])
    if conditions_met == 4:
        return "strong"
    elif conditions_met == 2:
        return "moderate"
    else:
        return "weak"
password_input = input("Enter your password: ")
print("Password is:", valid_password(password_input))
```

```python
def valid_password(password):
    has_uppercase = any(char.isupper() for char in password)
    has_number = any(char.isdigit() for char in password)
    has_special_char = any(not char.isalnum() for char in password)
    is_long_enough = len(password) >= 8

    conditions_met = sum([
        has_uppercase,
        has_number,
        has_special_char,
        is_long_enough
    ])

    if conditions_met == 4:
        return "strong"
    elif conditions_met == 2:
        return "moderate"
    else:
        return "weak"
user_password = input("Enter your password: ")
strength = valid_password(user_password)
print(f"Your password strength is: {strength}")
```
