```python
menu = {
    "tea": 120,
    "coffee": 200,
    "sandwich": 450,
    "cookie": 150,
    "cake_slice": 320
}

inventory = {
    "tea": 40,
    "coffee": 35,
    "sandwich": 20,
    "cookie": 50,
    "cake_slice": 25
}

orders = [
    {"tea": 2, "cookie": 1},
    {"coffee": 1, "sandwich": 2},
    {"tea": 3, "cake_slice": 1},
    {"coffee": 2, "cookie": 2, "sandwich": 1},
    {"tea": 1},
    {"coffee": 1, "cake_slice": 2}
]

# 1. VALIDATE ORDERS

processed_orders = []

for order in orders:
    cleaned_order = {}

    for item in order:
        quantity = order[item]
        if item not in menu:
            continue

        available_qty = inventory[item]
        if quantity > available_qty:
            quantity = available_qty

        if quantity > 0:
            cleaned_order[item] = quantity

    processed_orders.append(cleaned_order)

print("Processed Orders:")
for o in processed_orders:
    print(o)
```

```python
order_totals = []

for order in processed_orders:
    total = 0
    for item in order:
        quantity = order[item]
        price = menu[item]
        total += quantity * price
    order_totals.append(total)

print("Order Totals:")
for t in order_totals:
    print(t)
```

```python
for order in processed_orders:
    for item in order:
        quantity = order[item]
        inventory[item] -= quantity
```

```python
total_revenue = sum(order_totals)

items_sold = {}

for order in processed_orders:
    for item in order:
        qty = order[item]
        if item not in items_sold:
            items_sold[item] = qty
        else:
            items_sold[item] += qty

top_selling_item = None
highest_qty = 0

for item in items_sold:
    if items_sold[item] > highest_qty:
        highest_qty = items_sold[item]
        top_selling_item = item

print("----- DAY SUMMARY -----")
print("Total Revenue:", total_revenue)
print("\nItems Sold:")
for item, qty in items_sold.items():
    print(f"{item}: {qty}")

print("\nTop Selling Item:", top_selling_item)
```

```python
low_stock = []

for item in inventory:
    if inventory[item] <= 5:
        low_stock.append(item)

print("----- LOW STOCK ITEMS -----")
if len(low_stock) == 0:
    print("No items are low on stock.")
else:
    for item in low_stock:
        print(item)
```

```python
order_totals_with_discount = []

for total in order_totals:
    if total > 1000:
        discounted_total = total * 0.90
    else:
        discounted_total = total

    order_totals_with_discount.append(discounted_total)

total_revenue_after_discount = sum(order_totals_with_discount)

print("Order Totals With Discount:")
for t in order_totals_with_discount:
    print(t)

print("\nTotal Revenue After Discount:", total_revenue_after_discount)
```

```python
order_points = []

for total in order_totals_with_discount:
    points = int(total // 100)
    order_points.append(points)
total_points = sum(order_points)

print("Order Loyalty Points:")
for p in order_points:
    print(p)

print("\nTotal Loyalty Points Given Today:", total_points)
```

```python
# --------------------------------------
# GIVEN DATA
# --------------------------------------
menu = {
    "tea": 120,
    "coffee": 200,
    "sandwich": 450,
    "cookie": 150,
    "cake_slice": 320
}

inventory = {
    "tea": 40,
    "coffee": 35,
    "sandwich": 20,
    "cookie": 50,
    "cake_slice": 25
}

orders = [
    {"tea": 2, "cookie": 1},
    {"coffee": 1, "sandwich": 2},
    {"tea": 3, "cake_slice": 1},
    {"coffee": 2, "cookie": 2, "sandwich": 1},
    {"tea": 1},
    {"coffee": 1, "cake_slice": 2}
]

# --------------------------------------
# STEP 1: VALIDATE ORDERS
# --------------------------------------
processed_orders = []

for order in orders:
    cleaned_order = {}

    for item in order:
        if item not in menu:
            continue

        quantity = order[item]
        available = inventory[item]

        if quantity > available:
            quantity = available

        if quantity > 0:
            cleaned_order[item] = quantity

    processed_orders.append(cleaned_order)


# --------------------------------------
# STEP 2: CALCULATE BILL FOR EACH ORDER
# --------------------------------------
order_totals = []

for order in processed_orders:
    total = 0
    for item in order:
        total += order[item] * menu[item]
    order_totals.append(total)


# --------------------------------------
# STEP 3: UPDATE INVENTORY
# --------------------------------------
for order in processed_orders:
    for item in order:
        inventory[item] -= order[item]


# --------------------------------------
# STEP 4: DAY SUMMARY (REVENUE + ITEMS SOLD)
# --------------------------------------
total_revenue = sum(order_totals)

items_sold = {}

for order in processed_orders:
    for item in order:
        if item not in items_sold:
            items_sold[item] = order[item]
        else:
            items_sold[item] += order[item]

top_selling_item = max(items_sold, key=items_sold.get)


# --------------------------------------
# STEP 5: LOW STOCK REPORT
# --------------------------------------
low_stock = [item for item in inventory if inventory[item] <= 5]


# --------------------------------------
# STEP 6: DISCOUNT CALCULATIONS
# --------------------------------------
order_totals_with_discount = []

for total in order_totals:
    if total > 1000:
        total = total * 0.90  # Apply 10% discount
    order_totals_with_discount.append(total)

total_revenue_after_discount = sum(order_totals_with_discount)


# --------------------------------------
# STEP 7: LOYALTY POINTS
# --------------------------------------
order_points = []

for total in order_totals_with_discount:
    points = int(total // 100)
    order_points.append(points)

total_points = sum(order_points)


# --------------------------------------
# FINAL OUTPUT DISPLAY
# --------------------------------------

print("\n========= FINAL OUTPUT =========\n")

print("Processed Orders:")
for o in processed_orders:
    print(o)

print("\nOrder Totals (before discount):", order_totals)
print("Order Totals (after discount):", order_totals_with_discount)

print("\nTotal Revenue Before Discount:", total_revenue)
print("Total Revenue After Discount:", total_revenue_after_discount)

print("\nItems Sold:", items_sold)
print("Top Selling Item:", top_selling_item)

print("\nRemaining Inventory:", inventory)

print("\nLow Stock Items:", low_stock)

print("\nOrder Loyalty Points:", order_points)
print("Total Loyalty Points Given Today:", total_points)

print("\n========= END OF REPORT =========")
```
